```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, RobustScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from datetime import datetime
import seaborn as sns
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test.csv")
```

```python
def prepare_data(train_df, test_df):
    train = train_df.copy()
    test = test_df.copy()

    for df in [train, test]:
        # Add 2000 to year values
        if df['Year'].max() < 10:
            year_values = 2000 + df['Year']
        else:
            year_values = df['Year']

        df['date'] = pd.to_datetime({
            'year': year_values,
            'month': df['Month'],
            'day': df['Day']
        }, errors='coerce')

        # Drop rows where 'Day' is 29 and month is 2 (leap day issues)
        df.drop(df[(df['Month'] == 2) & (df['Day'] == 29)].index, inplace=True)

        # Check for NaT values
        if df['date'].isna().sum() > 0:
            print(f"Found {df['date'].isna().sum()} NaT values in date column")
            df.dropna(subset=['date'], inplace=True)

    train = train.drop(columns=['Year', 'Month', 'Day'])
    test = test.drop(columns=['Year', 'Month', 'Day'])

    train = train.set_index('date').sort_index()
    test = test.set_index('date').sort_index()

    return train, test
```

```python
train, test = prepare_data(train_df, test_df)
```

Check for missing values in both datasets


```python
print("Missing values in train:", train.isna().sum())
print("Missing values in test:", test.isna().sum())
```

Data cleaning and preprocessing


```python
def clean_data(df):
    df_cleaned = df.copy()

    df_cleaned = df_cleaned.fillna(method='ffill').fillna(method='bfill')

    if df_cleaned.isna().sum().sum() > 0:
        df_cleaned = df_cleaned.fillna(df_cleaned.median())

    # Kelvin to cels
    temp_columns = ['Avg_Temperature', 'Avg_Feels_Like_Temperature']
    KELVIN_THRESHOLD = 100

    for col in temp_columns:
        if col in df_cleaned.columns:
            df_cleaned[col] = df_cleaned[col].apply(lambda x: x - 273.15 if x > KELVIN_THRESHOLD else x)
            df_cleaned[col] = df_cleaned[col].round(2)

    rain_cols = ['Rain_Amount', 'Rain_Duration']
    for col in rain_cols:
        if col in df_cleaned.columns:
            neg_rain_mask = df_cleaned[col] < 0
            if neg_rain_mask.any():
                df_cleaned.loc[neg_rain_mask, col] = 0

    if 'Wind_Speed' in df_cleaned.columns:
        neg_wind_mask = df_cleaned['Wind_Speed'] < 0
        if neg_wind_mask.any():
            df_cleaned.loc[neg_wind_mask, 'Wind_Speed'] = df_cleaned.loc[neg_wind_mask, 'Wind_Speed'].abs()

    if 'Evapotranspiration' in df_cleaned.columns:
        neg_evap_mask = df_cleaned['Evapotranspiration'] < 0
        if neg_evap_mask.any():
            df_cleaned.loc[neg_evap_mask, 'Evapotranspiration'] = 0

    if 'Wind_Direction' in df_cleaned.columns:
        invalid_dir_mask = (df_cleaned['Wind_Direction'] < 0) | (df_cleaned['Wind_Direction'] > 360)
        if invalid_dir_mask.any():
            df_cleaned.loc[invalid_dir_mask, 'Wind_Direction'] = df_cleaned.loc[invalid_dir_mask, 'Wind_Direction'].apply(
                lambda x: (x % 360) if x != 0 else x
            )

    #  IQR method for Handle outliers
    for col in df_cleaned.select_dtypes(include='number').columns:
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 3 * IQR
        upper_bound = Q3 + 3 * IQR

        df_cleaned[col] = df_cleaned[col].clip(lower_bound, upper_bound)

    return df_cleaned
```

```python
train = clean_data(train)
test = clean_data(test)
```

Feature Engineering part

```python
def add_features(df):
    df_new = df.copy()

    df_new['month'] = df_new.index.month
    df_new['day'] = df_new.index.day
    df_new['day_of_year'] = df_new.index.dayofyear
    df_new['season'] = ((df_new.index.month % 12 + 3) // 3).map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})

    # sine and cosine transforms
    df_new['month_sin'] = np.sin(2 * np.pi * df_new['month']/12)
    df_new['month_cos'] = np.cos(2 * np.pi * df_new['month']/12)
    df_new['day_sin'] = np.sin(2 * np.pi * df_new['day']/31)
    df_new['day_cos'] = np.cos(2 * np.pi * df_new['day']/31)
    df_new['day_of_year_sin'] = np.sin(2 * np.pi * df_new['day_of_year']/365)
    df_new['day_of_year_cos'] = np.cos(2 * np.pi * df_new['day_of_year']/365)

    #doing One-hot encode for season
    season_dummies = pd.get_dummies(df_new['season'], prefix='season')
    df_new = pd.concat([df_new, season_dummies], axis=1)

    # Adding lag features for all target variables
    for lag in [1, 3, 7, 14]:
        for col in ['Avg_Temperature', 'Radiation', 'Rain_Amount', 'Wind_Speed', 'Wind_Direction']:
            if col in df_new.columns:
                df_new[f'{col}_lag_{lag}'] = df_new[col].shift(lag)

    # Add rolling statistics
    for window in [3, 7, 14]:
        for col in ['Avg_Temperature', 'Radiation', 'Rain_Amount', 'Wind_Speed']:
            if col in df_new.columns:
                df_new[f'{col}_rolling_mean_{window}'] = df_new[col].rolling(window=window, min_periods=1).mean()
                df_new[f'{col}_rolling_std_{window}'] = df_new[col].rolling(window=window, min_periods=1).std().fillna(0)

    # Add interaction terms between weather variables
    if all(col in df_new.columns for col in ['Avg_Temperature', 'Rain_Amount']):
        df_new['Temp_Rain_Interaction'] = df_new['Avg_Temperature'] * df_new['Rain_Amount']

    if all(col in df_new.columns for col in ['Wind_Speed', 'Rain_Amount']):
        df_new['Wind_Rain_Interaction'] = df_new['Wind_Speed'] * df_new['Rain_Amount']

    # Drop the categorical season column after encoding
    df_new = df_new.drop(columns=['season'])

    df_new = df_new.fillna(method='ffill').fillna(method='bfill').fillna(0)

    return df_new
```

```python
train_enhanced = add_features(train)
test_enhanced = add_features(test)
```

```python
test_enhanced.head()
```

```python
target_columns = [
    'Avg_Temperature',
    'Radiation',
    'Rain_Amount',
    'Wind_Speed',
    'Wind_Direction'
]
```

```python
actual_target_columns = [col for col in target_columns if col in train.columns]
print(f"Predicting these environmental variables: {actual_target_columns}")
```

 Calculating correlations for feature selection


```python
corr_matrix = train_enhanced.select_dtypes(include="number").corr()
corr_matrix
```

 Finding most important features for each target

```python
important_features = {}
for target in actual_target_columns:
    if target in corr_matrix.columns:

        corr_with_target = corr_matrix[target].abs().sort_values(ascending=False)

        top_features = corr_with_target[1:16].index.tolist()
        important_features[target] = top_features
```

 Finding union of all important features across targets

```python
all_important_features = set()
for features in important_features.values():
    all_important_features.update(features)
print(f"Using {len(all_important_features)} important features for prediction")
```

 Create sequences with improved sliding window approach

```python
def create_sequences(data, target_data, seq_length=30):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(target_data[i+seq_length])
    return np.array(X), np.array(y)
```

```python
features = train_enhanced[list(all_important_features) + ['month_sin', 'month_cos', 'day_sin', 'day_cos']]
```

check whether include all necessary columns from train_enhanced

```python

for col in actual_target_columns:
    if col not in features.columns:
        features[col] = train_enhanced[col]
```

```python
targets = train_enhanced[actual_target_columns]
targets
```

 Normalize the data

```python
feature_scaler = RobustScaler()
target_scaler = MinMaxScaler(feature_range=(0, 1))

scaled_features = feature_scaler.fit_transform(features)
scaled_targets = target_scaler.fit_transform(targets)
```

Creating sequences with optimized window length

```python
seq_length = 30
X, y = create_sequences(scaled_features, scaled_targets, seq_length)
```

Spliting into train and validation sets

```python
train_size = int(len(X) * 0.8)
X_train, X_val = X[:train_size], X[train_size:]
y_train, y_val = y[:train_size], y[train_size:]
```

Reshape for LSTM [samples, time steps, features]


```python
num_features = scaled_features.shape[1]
num_targets = len(actual_target_columns)
```

Creating Bidirectional LSTM model for better capturing of temporal dependencies

```python

def create_model(seq_length, num_features, num_targets):
    model = Sequential()


    model.add(Bidirectional(LSTM(128, return_sequences=True), input_shape=(seq_length, num_features)))
    model.add(Dropout(0.3))

    model.add(Bidirectional(LSTM(96, return_sequences=True)))
    model.add(Dropout(0.3))

    model.add(LSTM(64))
    model.add(Dropout(0.2))

    model.add(Dense(48, activation='relu'))
    model.add(Dense(num_targets))

    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
    model.compile(optimizer=optimizer, loss='huber')

    return model
```

Create and train the model


```python
model = create_model(seq_length, num_features, num_targets)
```



Set up callbacks

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=15,
    min_lr=0.0001,
    verbose=1
)
```

Add early stopping to training the model

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=70,
    batch_size=32,
    shuffle=True,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)
```

Traing loss and Validation loss plot

```python
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)
plt.show()
```

 calculating validation matrice (SMAPE,RMSE and MAE)


```python
def smape(actual, predicted):
    denominator = (np.abs(actual) + np.abs(predicted))

    denominator = np.where(denominator == 0, 1e-10, denominator)

    return 100/len(actual) * np.sum(2 * np.abs(predicted - actual) / denominator)

val_pred = model.predict(X_val)
val_pred_actual = target_scaler.inverse_transform(val_pred)
val_actual = target_scaler.inverse_transform(y_val)

for i, col in enumerate(actual_target_columns):
    val_smape = smape(val_actual[:, i], val_pred_actual[:, i])
    val_rmse = np.sqrt(mean_squared_error(val_actual[:, i], val_pred_actual[:, i]))
    val_mae = mean_absolute_error(val_actual[:, i], val_pred_actual[:, i])
    print(f"Validation metrics for {col}:")
    print(f"  - SMAPE: {val_smape:.2f}%")
    print(f"  - RMSE: {val_rmse:.2f}")
    print(f"  - MAE: {val_mae:.2f}")
```

predictions vs actual for validation set

```python
plt.figure(figsize=(18, 15))
for i, col in enumerate(actual_target_columns):
    plt.subplot(len(actual_target_columns), 1, i+1)
    plt.plot(val_actual[:100, i], label=f'Actual {col}')
    plt.plot(val_pred_actual[:100, i], label=f'Predicted {col}')
    plt.title(f'Validation Set: Actual vs Predicted {col}')
    plt.xlabel('Time Steps')
    plt.ylabel(col)
    plt.legend()
    plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
test_features = test_enhanced.copy()
```

Adding missing columns to test data


```python
for col in features.columns:
    if col not in test_features.columns:
        test_features[col] = 0
```

```python
test_features = test_features[features.columns]
```

Scale test features


```python
scaled_test_features = feature_scaler.transform(test_features)
```

 walk-forward approach


```python
if 'ID' in test.columns or 'ID' in test.reset_index().columns:
    last_sequence = scaled_features[-seq_length:]

    all_predictions = {col: [] for col in actual_target_columns}

    if 'ID' in test.columns:
        all_predictions['ID'] = test['ID'].values
    elif 'ID' in test.reset_index().columns:
        all_predictions['ID'] = test.reset_index()['ID'].values
    else:
        all_predictions['ID'] = list(range(len(test)))

    for i in range(len(test)):
        sequence = last_sequence.reshape(1, seq_length, num_features)

        predicted_scaled = model.predict(sequence, verbose=0)

        predicted_values = target_scaler.inverse_transform(predicted_scaled)[0]

        for j, col in enumerate(actual_target_columns):
            # Apply physical constraints
            if col == 'Rain_Amount' and predicted_values[j] < 0:
                predicted_values[j] = 0
            elif col == 'Wind_Speed' and predicted_values[j] < 0:
                predicted_values[j] = 0
            elif col == 'Wind_Direction':
                predicted_values[j] = predicted_values[j] % 360

            all_predictions[col].append(predicted_values[j])

        # Update sequence for next prediction using walk-forward validation
        if i < len(test) - 1:
            last_sequence = np.roll(last_sequence, -1, axis=0)

            last_sequence[-1] = scaled_test_features[i]

    test_predictions_df = pd.DataFrame(all_predictions)

    # Apply post-processing to ensure physical constraints
    if 'Wind_Direction' in test_predictions_df.columns:
       test_predictions_df['Wind_Direction'] = test_predictions_df['Wind_Direction'].apply(lambda x: x % 360)

    if 'Rain_Amount' in test_predictions_df.columns:
       test_predictions_df['Rain_Amount'] = test_predictions_df['Rain_Amount'].apply(lambda x: max(0, x))

    if 'Wind_Speed' in test_predictions_df.columns:
       test_predictions_df['Wind_Speed'] = test_predictions_df['Wind_Speed'].apply(lambda x: max(0, x))

    # Apply additional smoothing to reduce unrealistic fluctuations
    for col in actual_target_columns:
       if col != 'Wind_Direction':
           window_size = 3
           test_predictions_df[col] = test_predictions_df[col].rolling(window=window_size, min_periods=1, center=True).mean()

    submission_df = pd.DataFrame({
        'ID': test_predictions_df['ID'],
    })

    for col in actual_target_columns:
        submission_df[col] = test_predictions_df[col].round(2)
```

Visualize final predictions

```python
plt.figure(figsize=(16, 10))
for i, col in enumerate(actual_target_columns):
    plt.subplot(len(actual_target_columns), 1, i+1)
    plt.plot(submission_df[col].values, label=f'Predicted {col}')
    plt.title(f'Final Predictions: {col}')
    plt.xlabel('Test Data Index')
    plt.ylabel(col)
    plt.grid(True)
plt.tight_layout()
plt.show()
```
